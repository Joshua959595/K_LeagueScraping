# 2. 셀레니움 기능함수

```python
# 데이터 센터 클릭
def ClickDC(driver):
    loca = driver.find_element(By.XPATH, '/html/body/div[1]/div[1]/div/div[1]/div[2]/ul/li[1]/a')
    loca.click()
    t.sleep(4)

# 플레이어 스탯 클릭
def ClickPS(driver):
    loca = driver.find_element(By.XPATH, '/html/body/div[1]/div[6]/div[1]/div[2]/div/div/ul/li[3]/a')
    loca.click()
    t.sleep(4)
    
# 매치 리스트 클릭
def ClickMatchList(driver):
    loca = driver.find_element(By.ID, 'selGameId')
    loca.click()
    t.sleep(4)

# 라운드 리스트 클릭
def ClickRoundList(driver):
    loca = driver.find_element(By.ID, 'selRoundId')
    loca.click()
    t.sleep(2)

# 리그 리스트 클릭
def ClickLeagueList(driver):
    loca = driver.find_element(By.ID, 'selMeetSeq')
    loca.click()
    t.sleep(2)

# 연도 리스트 클릭
def ClickYearList(driver):
    loca = driver.find_element(By.ID, 'selMeetYear')
    loca.click()
    t.sleep(2)
    
## 해당 Path 객체 클릭
def ClickPathObj(driver, path):
    loca = driver.find_element(By.XPATH, path)
    loca.click()
    t.sleep(4)
    
def getSoup(driver):
    html = driver.page_source
    soup = bs(html, 'html.parser')
    t.sleep(2)
    return soup
```
