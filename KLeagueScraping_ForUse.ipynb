```python
# 데이터 센터 클릭
def ClickDC(driver):
    loca = driver.find_element(By.XPATH, '/html/body/div[1]/div[1]/div/div[1]/div[2]/ul/li[1]/a')
    loca.click()
    t.sleep(4)

# 플레이어 스탯 클릭
def ClickPS(driver):
    loca = driver.find_element(By.XPATH, '/html/body/div[1]/div[6]/div[1]/div[2]/div/div/ul/li[3]/a')
    loca.click()
    t.sleep(4)

    
## 해당 Path 객체 클릭
def ClickPathObj(driver, path):
    loca = driver.find_element(By.XPATH, path)
    loca.click()
    t.sleep(4)
    
# 매치 리스트 클릭
def ClickMatchList(driver):
    loca = driver.find_element(By.ID, 'selGameId')
    loca.click()
    t.sleep(4)

# 라운드 리스트 클릭
def ClickRoundList(driver):
    loca = driver.find_element(By.ID, 'selRoundId')
    loca.click()
    t.sleep(2)

# 리그 리스트 클릭
def ClickLeagueList(driver):
    loca = driver.find_element(By.ID, 'selMeetSeq')
    loca.click()
    t.sleep(2)

# 연도 리스트 클릭
def ClickYearList(driver):
    loca = driver.find_element(By.ID, 'selMeetYear')
    loca.click()
    t.sleep(2)
    
def getSoup(driver):
    html = driver.page_source
    soup = bs(html, 'html.parser')
    t.sleep(2)
    return soup
    
```

```python
def getAttInfo(soup):
    global HM, AM
    HM = 0
    AM = 0
    ### 홈팀
    hatt = soup.findAll("table", { "id" : "homeAttList" })
    hatt = str(hatt)
    hatt = bs(hatt, 'html.parser')
    hatt_trs = hatt.findAll('tr')
    for hatt_tr in hatt_trs[2:]: # 객체20개 (앞에 두개 안 씀)
        c = hatt_tr.text
        c = c.replace('\n', ',')
        c = c[1:]
        c = c.split(',')
       
        try:
            C_Backnum.append(c[0])
        except: 
            C_Backnum.append('Missing')
        try:
            C_Name.append(c[1])
        except:
            C_Name.append('Missing')
        try:
            C_Pos.append(c[2])
        except:
            C_Pos.append('Missing')
        try:
            C_Time.append(c[3])
        except:
            C_Time.append('Missing')
        try:
            C_Rate.append(c[4])
        except:
            C_Rate.append('Missing')
        try:    
            A_Score.append(c[5])
        except:
            A_Score.append('Missing')
        try:    
            A_Assist.append(c[6])
        except:
            A_Assist.append('Missing')
        try:
            A_Shoot.append(c[7])
        except:
            A_Shoot.append('Missing')
        try:    
            A_VShoot.append(c[8])
        except:
            A_VShoot.append('Missing')
        try:
            A_BShoot.append(c[9])
        except:
            A_BShoot.append('Missing')
        try:
            A_nonVShoot.append(c[10])
        except:
            A_nonVShoot.append('Missing')
        try:
            A_PAShoot.append(c[11])
        except:
            A_PAShoot.append('Missing')
        try:
            A_nonPAShoot.append(c[12])
        except:
            A_nonPAShoot.append('Missing')
        try:
            A_Off.append(c[13])
        except:
            A_Off.append('Missing')
        try:
            A_Free.append(c[14])
        except:
            A_Free.append('Missing')
        try:
            A_Corner.append(c[15])
        except:
            A_Corner.append('Missing')
        try:
            A_Throw.append(c[16])
        except:
            A_Throw.append('Missing')
        try:
            A_DribSucNum.append(c[17])
        except:
            A_DribSucNum.append('Missing')
        try:
            A_DribSucPer.append(c[18])
        except:
            A_DribSucPer.append('Missing')
        HM += 1
    
    ### 어웨이팀
    aatt = soup.findAll("table", { "id" : "awayAttList" })
    aatt = str(aatt)
    aatt = bs(aatt, 'html.parser')
    aatt_trs = aatt.findAll('tr')
    
    for aatt_tr in aatt_trs[2:]: 
        c = aatt_tr.text
        c = c.replace('\n', ',')
        c = c[1:]
        c = c.split(',')
       
        try:
            C_Backnum.append(c[0])
        except: 
            C_Backnum.append('Missing')
        try:
            C_Name.append(c[1])
        except:
            C_Name.append('Missing')
        try:
            C_Pos.append(c[2])
        except:
            C_Pos.append('Missing')
        try:
            C_Time.append(c[3])
        except:
            C_Time.append('Missing')
        try:
            C_Rate.append(c[4])
        except:
            C_Rate.append('Missing')
        try:    
            A_Score.append(c[5])
        except:
            A_Score.append('Missing')
        try:    
            A_Assist.append(c[6])
        except:
            A_Assist.append('Missing')
        try:
            A_Shoot.append(c[7])
        except:
            A_Shoot.append('Missing')
        try:    
            A_VShoot.append(c[8])
        except:
            A_VShoot.append('Missing')
        try:
            A_BShoot.append(c[9])
        except:
            A_BShoot.append('Missing')
        try:
            A_nonVShoot.append(c[10])
        except:
            A_nonVShoot.append('Missing')
        try:
            A_PAShoot.append(c[11])
        except:
            A_PAShoot.append('Missing')
        try:
            A_nonPAShoot.append(c[12])
        except:
            A_nonPAShoot.append('Missing')
        try:
            A_Off.append(c[13])
        except:
            A_Off.append('Missing')
        try:
            A_Free.append(c[14])
        except:
            A_Free.append('Missing')
        try:
            A_Corner.append(c[15])
        except:
            A_Corner.append('Missing')
        try:
            A_Throw.append(c[16])
        except:
            A_Throw.append('Missing')
        try:
            A_DribSucNum.append(c[17])
        except:
            A_DribSucNum.append('Missing')
        try:
            A_DribSucPer.append(c[18])
        except:
            A_DribSucPer.append('Missing')
        AM += 1
        
def getPassInfo(soup):
    ### 홈 팀
    hpass = soup.findAll("table", { "id" : "homePassList" })
    hpass = str(hpass)
    hpass = bs(hpass, 'html.parser')
    hpass_trs = hpass.findAll('tr')
    for hpass_tr in hpass_trs[2:]:
        c = hpass_tr.text
        c = c.replace('\n', ',')
        c = c[1:] # 앞에 콤마 제거
        c = c.split(',')
        
        try:
            P_PassSucNum.append(c[5])
        except:
            P_PassSucNum.append('Missing')
        try:
            P_PassSucPer.append(c[6])
        except:
            P_PassSucPer.append('Missing')
        try:
            P_Key.append(c[7])
        except:
            append('Missing')
        try:
            P_AASucNum.append(c[8])
        except:
            P_AASucNum.append('Missing')
        try:
            P_AASucPer.append(c[9])
        except:
            P_AASucPer.append('Missing')
        try:
            P_MASucNum.append(c[10])
        except:
            P_MASucNum.append('Missing')
        try:
            P_MASucPer.append(c[11])
        except:
            P_MASucPer.append('Missing')
        try:
            P_DASucNum.append(c[12])
        except:
            P_DASucNum.append('Missing')
        try:
            P_DASucPer.append(c[13])
        except:
            P_DASucPer.append('Missing')
        try:
            P_LongSucNum.append(c[14])
        except:
            P_LongSucNum.append('Missing')
        try:
            P_LongSucPer.append(c[15])
        except:
            P_LongSucPer.append('Missing')
        try:
            P_MidSucNum.append(c[16])
        except:
            P_MidSucNum.append('Missing')
        try:
            P_MidSucPer.append(c[17])
        except:
            P_MidSucPer.append('Missing')
        try:
            P_ShortSucNum.append(c[18])
        except:
            P_ShortSucNum.append('Missing')
        try:
            P_ShortSucPer.append(c[19])
        except:
            P_ShortSucPer.append('Missing')
        try:
            P_ForSucNum.append(c[20])
        except:
            P_ForSucNum.append('Missing')
        try:
            P_ForSucPer.append(c[21])
        except:
            P_ForSucPer.append('Missing')
        try:
            P_SidSucNum.append(c[22])
        except:
            P_SidSucNum.append('Missing')
        try:
            P_SidSucPer.append(c[23])
        except:
            P_SidSucPer.append('Missing')
        try:
            P_BackSucNum.append(c[24])
        except:
            P_BackSucNum.append('Missing')
        try:
            P_BackSucPer.append(c[25])
        except:
            P_BackSucPer.append('Missing')
        try:
            P_CrossSucNum.append(c[26])
        except:
            P_CrossSucNum.append('Missing')
        try:
            P_CrossSucPer.append(c[27])
        except:
            P_CrossSucPer.append('Missing')
        try:
            P_PR.append(c[28])
        except:
            P_PR.append('Missing')
    
    ### 어웨이 팀
    apass = soup.findAll("table", { "id" : "awayPassList" })
    apass = str(apass)
    apass = bs(apass, 'html.parser')
    apass_trs = apass.findAll('tr')
    for apass_tr in apass_trs[2:]:
        c = apass_tr.text
        c = c.replace('\n', ',')
        c = c[1:] # 앞에 콤마 제거
        c = c.split(',')
       
        try:
            P_PassSucNum.append(c[5])
        except:
            P_PassSucNum.append('Missing')
        try:
            P_PassSucPer.append(c[6])
        except:
            P_PassSucPer.append('Missing')
        try:
            P_Key.append(c[7])
        except:
            append('Missing')
        try:
            P_AASucNum.append(c[8])
        except:
            P_AASucNum.append('Missing')
        try:
            P_AASucPer.append(c[9])
        except:
            P_AASucPer.append('Missing')
        try:
            P_MASucNum.append(c[10])
        except:
            P_MASucNum.append('Missing')
        try:
            P_MASucPer.append(c[11])
        except:
            P_MASucPer.append('Missing')
        try:
            P_DASucNum.append(c[12])
        except:
            P_DASucNum.append('Missing')
        try:
            P_DASucPer.append(c[13])
        except:
            P_DASucPer.append('Missing')
        try:
            P_LongSucNum.append(c[14])
        except:
            P_LongSucNum.append('Missing')
        try:
            P_LongSucPer.append(c[15])
        except:
            P_LongSucPer.append('Missing')
        try:
            P_MidSucNum.append(c[16])
        except:
            P_MidSucNum.append('Missing')
        try:
            P_MidSucPer.append(c[17])
        except:
            P_MidSucPer.append('Missing')
        try:
            P_ShortSucNum.append(c[18])
        except:
            P_ShortSucNum.append('Missing')
        try:
            P_ShortSucPer.append(c[19])
        except:
            P_ShortSucPer.append('Missing')
        try:
            P_ForSucNum.append(c[20])
        except:
            P_ForSucNum.append('Missing')
        try:
            P_ForSucPer.append(c[21])
        except:
            P_ForSucPer.append('Missing')
        try:
            P_SidSucNum.append(c[22])
        except:
            P_SidSucNum.append('Missing')
        try:
            P_SidSucPer.append(c[23])
        except:
            P_SidSucPer.append('Missing')
        try:
            P_BackSucNum.append(c[24])
        except:
            P_BackSucNum.append('Missing')
        try:
            P_BackSucPer.append(c[25])
        except:
            P_BackSucPer.append('Missing')
        try:
            P_CrossSucNum.append(c[26])
        except:
            P_CrossSucNum.append('Missing')
        try:
            P_CrossSucPer.append(c[27])
        except:
            P_CrossSucPer.append('Missing')
        try:
            P_PR.append(c[28])
        except:
            P_PR.append('Missing')
    
def getDefInfo(soup):
    ### 홈데이터
    hdef = soup.findAll("table", { "id" : "homeDefList" })
    hdef = str(hdef)
    hdef = bs(hdef, 'html.parser')
    hdef_trs = hdef.findAll('tr')
    for hdef_tr in hdef_trs[2:]:
        c = hdef_tr.text
        c = c.replace('\n', ',')
        c = c[1:] # 앞에 콤마 제거
        c = c.split(',')
        
        try:
            D_GroundSucNum.append(c[5])
        except:
            D_GroundSucNum.append('Missing')
        try:
            D_GroundSucPer.append(c[6])
        except:
            D_GroundSucPer.append('Missing')
        try:
            D_AirSucNum.append(c[7])
        except:
            D_AirSucNum.append('Missing')
        try:
            D_AirSucPer.append(c[8])
        except:
            D_AirSucPer.append('Missing')
        try:
            D_TackSucNum.append(c[9])
        except:
            D_TackSucNum.append('Missing')
        try:
            D_TackSucPer.append(c[10])
        except:
            D_TackSucPer.append('Missing')
        try:
            D_Clear.append(c[11])
        except:
            D_Clear.append('Missing')
        try:
            D_Inter.append(c[12])
        except:
            D_Inter.append('Missing')
        try:
            D_Shut.append(c[13])
        except:
            D_Shut.append('Missing')
        try:
            D_Gain.append(c[14])
        except:
            D_Gain.append('Missing')
        try:
            D_Block.append(c[15])
        except:
            D_Block.append('Missing')
        try:
            D_Miss.append(c[16])
        except:
            D_Miss.append('Missing')
        try:
            D_Foul.append(c[17])
        except:
            D_Foul.append('Missing')
        try:
            D_GainFoul.append(c[18])
        except:
            D_GainFoul.append('Missing')
        try:
            D_Yellow.append(c[19])
        except:
            D_Yellow.append('Missing')
        try:
            D_Red.append(c[20])
        except:
            D_Red.append('Missing')
    
    ### 어웨이데이터
    adef = soup.findAll("table", { "id" : "awayDefList" })
    adef = str(adef)
    adef = bs(adef, 'html.parser')
    adef_trs = adef.findAll('tr')
    for adef_tr in adef_trs[2:]:
        c = adef_tr.text
        c = c.replace('\n', ',')
        c = c[1:] # 앞에 콤마 제거
        c = c.split(',')
        
        try:
            D_GroundSucNum.append(c[5])
        except:
            D_GroundSucNum.append('Missing')
        try:
            D_GroundSucPer.append(c[6])
        except:
            D_GroundSucPer.append('Missing')
        try:
            D_AirSucNum.append(c[7])
        except:
            D_AirSucNum.append('Missing')
        try:
            D_AirSucPer.append(c[8])
        except:
            D_AirSucPer.append('Missing')
        try:
            D_TackSucNum.append(c[9])
        except:
            D_TackSucNum.append('Missing')
        try:
            D_TackSucPer.append(c[10])
        except:
            D_TackSucPer.append('Missing')
        try:
            D_Clear.append(c[11])
        except:
            D_Clear.append('Missing')
        try:
            D_Inter.append(c[12])
        except:
            D_Inter.append('Missing')
        try:
            D_Shut.append(c[13])
        except:
            D_Shut.append('Missing')
        try:
            D_Gain.append(c[14])
        except:
            D_Gain.append('Missing')
        try:
            D_Block.append(c[15])
        except:
            D_Block.append('Missing')
        try:
            D_Miss.append(c[16])
        except:
            D_Miss.append('Missing')
        try:
            D_Foul.append(c[17])
        except:
            D_Foul.append('Missing')
        try:
            D_GainFoul.append(c[18])
        except:
            D_GainFoul.append('Missing')
        try:
            D_Yellow.append(c[19])
        except:
            D_Yellow.append('Missing')
        try:
            D_Red.append(c[20])
        except:
            D_Red.append('Missing')
    
    
def getGkInfo(soup):
    ### 홈데이터
    hgk = soup.findAll("table", { "id" : "homeGkList" })
    hgk = str(hgk)
    hgk = bs(hgk, 'html.parser')
    hgk_trs = hgk.findAll('tr')
    count = 0 # 골키퍼 교체 출전 가능성 고려
    
    for hgk_tr in hgk_trs[2:]:

        c = hgk_tr.text
        c = c.replace('\n', ',')
        c = c[1:] # 앞에 콤마 제거
        c = c.split(',')
        
        try:
            G_Loss.append(c[5])
        except:
            G_Loss.append('Missing')
        try:
            G_Catch.append(c[6])
        except:
            G_Catch.append('Missing')
        try:
            G_Punch.append(c[7])
        except:
            G_Punch.append('Missing')
        try:
            G_KickSucNum.append(c[8])
        except:
            G_KickSumNum.append('Missing')
        try:
            G_KickSucPer.append(c[9])
        except:
            G_KickSucPer.append('Missing')
        try:
            G_AirSucNum.append(c[10])
        except:
            G_AirSucNum.append('Missing')
        try:
            G_AirSucPer.append(c[11])
        except:
            G_AirSucPer.append('Missing')
        count += 1
    for i in range(HM-count):
        G_Loss.append(None)
        G_Catch.append(None)
        G_Punch.append(None)
        G_KickSucNum.append(None)
        G_KickSucPer.append(None)
        G_AirSucNum.append(None)
        G_AirSucPer.append(None)

    
    ### 어웨이데이터
    agk = soup.findAll("table", { "id" : "awayGkList" })
    agk = str(agk)
    agk = bs(agk, 'html.parser')
    agk_trs = agk.findAll('tr')
    count = 0
    
    for agk_tr in agk_trs[2:]:
        c = agk_tr.text
        c = c.replace('\n', ',')
        c = c[1:] # 앞에 콤마 제거
        c = c.split(',')
        
        try:
            G_Loss.append(c[5])
        except:
            G_Loss.append('Missing')
        try:
            G_Catch.append(c[6])
        except:
            G_Catch.append('Missing')
        try:
            G_Punch.append(c[7])
        except:
            G_Punch.append('Missing')
        try:
            G_KickSucNum.append(c[8])
        except:
            G_KickSumNum.append('Missing')
        try:
            G_KickSucPer.append(c[9])
        except:
            G_KickSucPer.append('Missing')
        try:
            G_AirSucNum.append(c[10])
        except:
            G_AirSucNum.append('Missing')
        try:
            G_AirSucPer.append(c[11])
        except:
            G_AirSucPer.append('Missing')
        count += 1
    for i in range(AM-count):
        G_Loss.append(None)
        G_Catch.append(None)
        G_Punch.append(None)
        G_KickSucNum.append(None)
        G_KickSucPer.append(None)
        G_AirSucNum.append(None)
        G_AirSucPer.append(None)

        
def getBackInfo1(soup):
    
    # 팀이름/팀득점
    ## 홈
    div = soup.findAll("div", { "class" : "main-soccer-teamName flR pr20" })
    div = str(div) 
    div = bs(div, 'html.parser') 
    for i in range(HM):
        try:
            B_Team.append(div.find("span", { "class" : "main-soccer-teamName-txt pr10" }).text.strip()) # 팀이름
        except:
            B_Team.append('Missing')
        try:
            B_TeamScore.append(div.find("span", { "class" : "main-soccer-teamName-txt02" }).text.strip()) # 팀득점
        except:
            B_TeamScore.append('Missing')
    ## 어웨이
    div = soup.findAll("div", { "class" : "main-soccer-teamName flL pl20 border-right-none" })
    div = str(div) 
    div = bs(div, 'html.parser') 
    for i in range(AM):
        try:
            B_Team.append(div.find("span", { "class" : "main-soccer-teamName-txt01 pl10" }).text.strip()) # 어웨이팀이름
        except:
            B_Team.append('Missing')
        try:
            B_TeamScore.append(div.find("span", { "class" : "main-soccer-teamName-txt02" }).text.strip()) # 어웨이팀득점
        except:
            B_TeamScore.append('Missing')
        
    # 상대 팀이름
    for i in range(AM):
        try:
            B_OppTeam.append(div.find("span", { "class" : "main-soccer-teamName-txt01 pl10" }).text.strip())
        except:
            B_OppTeam.append('Missing')
    div = soup.findAll("div", { "class" : "main-soccer-teamName flR pr20" })
    div = str(div) 
    div = bs(div, 'html.parser') 
    for i in range(HM):
        try:
            B_OppTeam.append(div.find("span", { "class" : "main-soccer-teamName-txt pr10" }).text.strip())
        except:
            B_OppTeam.append('Missing')
    
    # 홈팀여부
    for i in range(HM+AM):
        if i < HM:
            B_isHome.append(1)
        else:
            B_isHome.append(0)
    
    # 감독
    # 홈팀
    div = soup.findAll("div", { "class" : "main-soccer-logo" })
    div = str(div) 
    div = bs(div, 'html.parser') 
    for i in range(HM):
        B_Coach.append(div.find("li", {"class" : "main-soccer-txt02 mt5"}).text.strip()) 
    # 어웨이팀
    div = soup.findAll("div", { "class" : "main-soccer-logo02" })
    div = str(div) 
    div = bs(div, 'html.parser') 
    for i in range(AM):
        B_Coach.append(div.find("li", {"class" : "main-soccer-txt02 mt5"}).text.strip()) 

def getBackInfo2(soup):
    divs = soup.findAll("div", { "class" : "match-prame-box" })
    divs = str(divs) 
    divs = bs(divs, 'html.parser') 

    # 경기장소
    for i in range(HM+AM):
        div = divs.findAll("li", { "class" : "match-prame-box-width01" })
        div = str(div) 
        div = bs(div, 'html.parser')
        try:
            B_Place.append(div.find("span", {"class" : "match-prame-boxTxt02"}).text.strip())
        except:
            B_Place.append('Missing')
            
    # 날짜
    for i in range(HM+AM):
        div = divs.find("li", { "class" : "match-prame-box-width04" })
        div = str(div)
        div = bs(div, 'html.parser')
        try:
            B_Date.append(div.find("span", {"class" : "match-prame-boxTxt02"}).text.strip())
        except:
            B_Date.append('Missing')
            
    # 관중 수
    for i in range(HM+AM):
        div = divs.find("li", { "class" : "match-prame-box-width02" })
        div = str(div)
        div = bs(div, 'html.parser')
        try:
            B_Crowd.append(div.find("span", {"class" : "match-prame-boxTxt02"}).text.strip())
        except:
            B_Crow.append('Missing')
    
    # 주심
    for i in range(HM+AM):
        div = divs.find("li", { "class" : "match-prame-box-width03" })
        div = str(div)
        div = bs(div, 'html.parser')
        try:
            B_Ref.append(div.find("span", {"class" : "match-prame-boxTxt02"}).text.strip())
        except:
            B_Ref.append('Missing')
            
    # 홈&어웨이 승리여부 
    global HS, AS
    HS = 0
    AS = 0
    ## 홈
    div = soup.findAll("div", { "class" : "main-soccer-teamName flR pr20" })
    div = str(div) 
    div = bs(div, 'html.parser') 
    HS = div.find("span", { "class" : "main-soccer-teamName-txt02" }).text.strip()
    ## 어웨이
    div = soup.findAll("div", { "class" : "main-soccer-teamName flL pl20 border-right-none" })
    div = str(div) 
    div = bs(div, 'html.parser')
    AS = div.find("span", { "class" : "main-soccer-teamName-txt02" }).text.strip()
    HS = int(HS)
    AS = int(AS)
    if (HS>AS):
        for i in range(HM+AM):
            B_HomeWin.append(1)
            B_AwayWin.append(0)
    elif (HS==AS):
        for i in range(HM+AM):
            B_HomeWin.append(0)
            B_AwayWin.append(0)
    else :
        for i in range(HM+AM):
            B_HomeWin.append(0)
            B_AwayWin.append(1)
    
        
```
